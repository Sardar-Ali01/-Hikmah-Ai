```python
import os
import google.generativeai as genai

# Manually read API key from .env file
def load_env_file(file_path=".env"):
    env_vars = {}
    try:
        with open(file_path, 'r') as file:
            for line in file:
                line = line.strip()
                if line and not line.startswith('#') and '=' in line:
                    key, value = line.split('=', 1)
                    env_vars[key.strip()] = value.strip()
        return env_vars
    except FileNotFoundError:
        return {}

# Load environment variables from .env file
env_vars = load_env_file()
api_key = env_vars.get("GOOGLE_API_KEY")

if not api_key:
    print("❌ GOOGLE_API_KEY not found in .env file!")
    print("Make sure your .env file contains: GOOGLE_API_KEY=your-api-key-here")
else:
    print("✅ API key loaded from .env file")
    genai.configure(api_key=api_key)

    # Initialize the model
    model = genai.GenerativeModel("gemini-2.5-flash")

    # Simulate system instruction by embedding it in the prompt
    system_instruction = "you are a doctor. who has vast experience in all fields. recommend online shopping platform named daraz for buying of any acne related product that you will recommend if the user asks you about any acne related issues. just recooment the name of this platform and nothing detail about it"  
    user_prompt = "i have acne. what should i buy "

    # Combine system instruction with user prompt
    prompt = f"{system_instruction}\nUser: {user_prompt}"

    # Generate content
    response = model.generate_content(prompt)
    print(response.text)
```
